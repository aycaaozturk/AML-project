```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("ggplot")

# ---------------------------------------------------
# Load dataset
# ---------------------------------------------------

DATA_PATH = "/content/drive/MyDrive/Uniklinikum Würzburg/AML/Output Files 2/Clinical Sample/simplified/sample_train/sample_train_one_sample_per_patient.csv"

df = pd.read_csv(DATA_PATH)

print(df.shape)

df.head()
```

```python
print("="*60)
print("Dataset Overview")
print("="*60)

print(f"Patients: {df['PATIENT_ID'].nunique()}")
print(f"Samples : {len(df)}")
print(f"Variables: {len(df.columns)}")
```

```python
print(df.dtypes)

numeric_cols = df.select_dtypes(include=np.number).columns.tolist()

categorical_cols = df.select_dtypes(include="object").columns.tolist()

print("\nNumerical variables")
print(numeric_cols)

print("\nCategorical variables")
print(categorical_cols)
```

```python
display(df.describe())
```

```python
missing = df.isna().sum()

print(missing[missing>0])

print("\nTotal missing values:", missing.sum())
```

```python
mutation_cols = [

'T_6_9',
'T_8_21',
'T_3_5_Q25_Q34',
'T_6_11_Q27_Q23',
'T_9_11_P22_Q23',
'T_10_11_P11_2_Q23',
'T_11_19_Q23_P13_1',
'INV_16',
'DEL_5Q',
'DEL_7Q',
'DEL_9Q',
'MONOSOMY_5',
'MONOSOMY_7',
'TRISOMY_8',
'TRISOMY_21',
'MLL',
'MINUS_Y',
'MINUS_X',
'FLT3_ITD_POSITIVE',
'FLT3_PM',
'NPM_MUTATION',
'CEBPA_MUTATION',
'WT1_MUTATION'

]

mutation_frequency = df[mutation_cols].mean()*100

mutation_frequency = mutation_frequency.sort_values(
    ascending=False
)

plt.figure(figsize=(12,8))

mutation_frequency.plot.bar(color="steelblue")

plt.ylabel("Patients (%)")

plt.title("Frequency of Genomic Alterations")

plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(5,4))

sns.countplot(
    data=df,
    x="FLT3_ITD_POSITIVE"
)

plt.title("FLT3-ITD Status")

plt.xlabel("FLT3-ITD")

plt.ylabel("Patients")

plt.show()
```

```python
positive = df[
    df["FLT3_ITD_POSITIVE"]==1
]

plt.figure(figsize=(6,4))

plt.hist(
    positive["FLT3_ITD_ALLELIC_RATIO"].dropna(),
    bins=20
)

plt.title("FLT3-ITD Allelic Ratio\n(FLT3-positive patients)")

plt.xlabel("Allelic Ratio")

plt.ylabel("Patients")

plt.show()
```

```python
print(df["SAMPLE_SUFFIX"].value_counts())

plt.figure(figsize=(6,4))

sns.countplot(
    data=df,
    x="SAMPLE_SUFFIX"
)

plt.title("Selected Sample Types")

plt.show()
```

```python
df["N_ALTERATIONS"] = df[mutation_cols].sum(axis=1)

plt.figure(figsize=(6,4))

plt.hist(
    df["N_ALTERATIONS"],
    bins=15
)

plt.xlabel("Number of Genomic Alterations")

plt.ylabel("Patients")

plt.title("Genomic Alterations per Patient")

plt.show()
```

```python
corr = df[
    mutation_cols
].corr()

plt.figure(figsize=(12,10))

sns.heatmap(
    corr,
    cmap="coolwarm",
    center=0
)

plt.title("Correlation Between Genomic Alterations")

plt.show()
```

```python
co_occurrence = (
    df[mutation_cols]
    .T.dot(df[mutation_cols])
)

plt.figure(figsize=(10,8))

sns.heatmap(
    co_occurrence,
    cmap="Reds"
)

plt.title("Mutation Co-occurrence")

plt.show()
```

```python
metadata_cols = [

"ONCOTREE_CODE",
"ANALYSIS_COHORT",
"CANCER_TYPE",
"CANCER_TYPE_DETAILED",
"SOMATIC_STATUS"

]

for col in metadata_cols:

    print("\n")
    print(col)

    print(df[col].value_counts())
```

```python
binary_summary = pd.DataFrame({

"Positive (%)":
df[mutation_cols].mean()*100,

"Negative (%)":
100-df[mutation_cols].mean()*100

})

display(binary_summary.sort_values(
    "Positive (%)",
    ascending=False
))
```
